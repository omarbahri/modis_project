```python
# Import packages
import os
import warnings

import matplotlib.pyplot as plt
import numpy.ma as ma
import xarray as xr
import rioxarray as rxr
#from shapely.geometry import mapping, box
#import geopandas as gpd
#import earthpy as et
#import earthpy.plot as ep
import numpy as np
from glob import glob

warnings.simplefilter('ignore')
```

```python
def combine_tifs(tif_list):
    """A function that combines a list of tifs in the same CRS
    and of the same extent into an xarray object

    Parameters
    ----------
    tif_list : list
        A list of paths to the tif files that you wish to combine.

    Returns
    -------
    An xarray object with all of the tif files in the listmerged into 
    a single object.

    """

    out_xr = []
    for i, tif_path in enumerate(tif_list):
        out_xr.append(rxr.open_rasterio(tif_path, masked=True).squeeze())
        out_xr[i]["band"] = i+1

    return xr.concat(out_xr, dim="band")
```

```python
data_path = os.path.join('/home/omar/Data/MODIS/Pecs/')
file_name = 'pecs_mod09_04_25_09_20.hdf'
```

```python
# Open just the bands that you want to process and masking the no-data values
desired_bands = ["500m Surface Reflectance Band 1", "500m Surface Reflectance Band 2", "500m Surface Reflectance Band 3",
                "500m Surface Reflectance Band 4", "500m Surface Reflectance Band 5", "500m Surface Reflectance Band 6",
                "500m Surface Reflectance Band 7"]
```

```python
# Notice that here, you get a single xarray object with just the bands that
# you want to work with
modis_bands = rxr.open_rasterio(os.path.join(data_path,file_name),
                                    masked=True,
                                    variable=desired_bands).squeeze()
```

```python
modis_bands["500m Surface Reflectance Band 1"]
```

```python
#Converting and saving all the bands to tif files
for i in range(7):
    output_tif_path = os.path.join(data_path,
                                     "tifs",
                                     "band_" + str(i+1)+ ".tif")
    
    name = "500m Surface Reflectance Band " + str(i+1)
    
    modis_bands[name].astype(np.int16).rio.to_raster(output_tif_path)
```

```python
for i in range(7):
    stacked_file_path = os.path.join(data_path,
                                     "tifs",
                                     "band_" + str(i+1)+ ".tif")
    
    name = "sur_refl_b0" + str(i+1) + "_1"
    
    modis_bands[name].astype(np.int16).rio.to_raster(stacked_file_path)
```

```python
#Load them again    
modis_bands_list = glob(os.path.join(data_path,
                                         "tifs",
                                         "*.tif"))

# Sort the list of bands
modis_bands_list.sort()
```

```python
# Open file list with function
modis_bands = combine_tifs(modis_bands_list)
```
